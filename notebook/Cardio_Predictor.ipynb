```python
#select only numerical features and drop the 'id' column
x=data.select_dtypes(include=["int64","float64"]).drop(columns=["num","target", "id"])
y=data["target"]
```

```python
#replacing the missing value with mean not median because medical data are normally distributed
x.fillna(x.mean(),inplace=True)
```

```python
#we will now split training and testing sets,we will use random state which can be any integer i have taken 42 so that no matter how many times i execute my code the result would be same
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Feature Scaling :this is important for logistic regreesion as it ensures faster convergence of gradient descent and prevents features with large scale to dominate learning process
from sklearn.preprocessing import StandardScaler

scaler =StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#importing logistic regression model from sklearn
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#predict class labels  computes probability using sigmoid it helps to converts probability  into class
y_pred=model.predict(x_test)
```

```python
#predict probabilites of positive class (class 1)
y_prob=model.predict_proba(x_test)[:,1]
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
```
